```python
from getpass import getpass
import instaloader
import pandas as pd
from instaloader import Post
from instaloader.exceptions import BadResponseException, ConnectionException, LoginRequiredException, TooManyRequestsException
import time
from IPython.display import display, clear_output


# Function to log in to Instagram
def login_instagram(username, password):
    L = instaloader.Instaloader()
    try:
        L.login(username, password)
        print("Logged in to Instagram")
        return L
    except (BadResponseException, ConnectionException, LoginRequiredException) as e:
        print(f"Error during login: {e}")
        return None

# Function to scrape Instagram post data
def scrape_instagram_post(username, password, post_link):
    L = login_instagram(username, password)
    if L is None:
        return None

    try:
        start_time = time.time()

        # Extract shortcode from the post link
        shortcode = post_link.split("/")[-2]

        # Load the post using the shortcode
        post = Post.from_shortcode(L.context, shortcode)

        # Initialize dictionary to hold the post data
        data = {
            "post_id": post.mediaid,
            "post_shortcode": post.shortcode,
            "post_date": post.date,
            "post_caption": post.caption,
            "post_likes": post.likes,
            "image_url": post.url,
            "post_is_video": post.is_video,
            "post_hashtags": post.caption_hashtags,
            "post_mentions": post.caption_mentions,
            "comments": []
        }

        # Get the total number of comments
        total_comments = post.comments
        post_likes = post.likes
        print(f"The post has {total_comments} comments and {post_likes} likes.")

        # Iterate over comments
        for i, comment in enumerate(post.get_comments(), start=1):
          if i%5 == 0:
            print(f"Scraping comment {i} of {total_comments}")
          data["comments"].append(comment.text)

        # Convert the data to a DataFrame
        df = pd.DataFrame([data])

        end_time = time.time()
        time_taken = end_time - start_time
        print(f"Time taken to scrape the post: {time_taken:.2f} seconds")

        return df

    except (BadResponseException, ConnectionException, TooManyRequestsException) as e:
        print(f"Error fetching post data: {e}")
        return None

# Main function to execute the script
def main():
    # Get user input
    username = getpass('Enter your Instagram username: ')
    password = getpass('Enter your Instagram password: ')
    post_link = input('Enter the link to the Instagram post: ')

    # Scrape the post data
    df = scrape_instagram_post(username, password, post_link)

    # Display the DataFrame if fetched successfully
    if df is not None:
        # Set display options
        pd.set_option('display.width', 500)
        pd.set_option('display.max_colwidth', 100)

        display(df)

if __name__ == "__main__":
    main()
```
