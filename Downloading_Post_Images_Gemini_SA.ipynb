```python
def get_avg_metrics(brand_name):
    """
    Retrieves the average likes, comments, and activity for a given brand name.

    Args:
    - brand_name (str): The name of the brand to retrieve average metrics for.

    Returns:
    - tuple: A tuple containing average likes, average comments, and average activity for the brand, converted to floats.
    - None: If the brand name is not found in the fashion_brands dictionary.
    """
    # Convert brand_name to lowercase
    brand_name = brand_name.lower()

    # Initialize a variable to store the found brand
    found_brand = None

    # Check if the brand_name matches any key in fashion_brands dictionary in a case-insensitive manner
    for brand in fashion_brands:
        if brand.lower().startswith(brand_name):
            found_brand = brand
            break

    if found_brand:
        # Get the average likes, comments, and activity for the brand
        avg_likes_str = fashion_brands[found_brand]["Stats"].get("Avg. likes", "0")
        avg_comments_str = fashion_brands[found_brand]["Stats"].get("Avg. comments", "0")

        # Convert average likes, comments, and activity strings to float
        def convert_to_float(value_str):
            if value_str[-1] == 'k':
                return float(value_str[:-1]) * 1000
            elif value_str[-1] == 'm':
                return float(value_str[:-1]) * 1000000
            else:
                return float(value_str)

        avg_likes = convert_to_float(avg_likes_str)
        avg_comments = convert_to_float(avg_comments_str)

        return avg_likes, avg_comments
    else:
        return None
```

```python
# Function to download post images using shortcodes and rename them correctly
def download_images(shortcode, post_id, target_folder):
    retries = 5
    for i in range(retries):
        try:
            post = instaloader.Post.from_shortcode(L.context, shortcode)
            print(f"Downloading post {shortcode} (ID: {post_id}) into folder '{target_folder}'")

            # Track image count
            image_count = 1

            # Check if the post is a sidecar (carousel)
            if post.typename == 'GraphSidecar':
                for item in post.get_sidecar_nodes():
                    image_url = item.display_url
                    L.download_pic(
                        filename=f"{target_folder}/{image_count:03d}_{post_id}",
                        url=image_url,
                        mtime=post.date_local
                    )
                    image_count += 1

            # For single image or video posts
            else:
                image_url = post.url
                L.download_pic(
                    filename=f"{target_folder}/{image_count:03d}_{post_id}",
                    url=image_url,
                    mtime=post.date_local
                )

            print(f"Successfully downloaded post {shortcode} (ID: {post_id})")
            break  # Break the loop if successful
        except Exception as e:
            print(f"Error downloading post {shortcode} (ID: {post_id}): {e}")
            print(f"Retrying ({i + 1}/{retries})...")
            time.sleep(2 ** i)  # Exponential backoff

# Function to process each CSV file
def process_csv_file(file_name):
    print(f"Processing file '{file_name}'")
    df = pd.read_csv(file_name)
    target_folder = os.path.splitext(file_name)[0]  # Folder name same as CSV file name
    os.makedirs(target_folder, exist_ok=True)  # Create folder if it doesn't exist
    print(f"Created target folder '{target_folder}'")

    unique_posts = df[['post_id', 'post_shortcode']].drop_duplicates()
    print(f"Found {len(unique_posts)} unique posts in '{file_name}'")

    for index, row in unique_posts.iterrows():
        post_id = row['post_id']
        shortcode = row['post_shortcode']
        print(f"Processing post {shortcode} (ID: {post_id})")
        download_images(shortcode, post_id, target_folder)
        print(f"Waiting for 10 seconds to avoid rate limits...")
        time.sleep(10)  # Delay to avoid hitting rate limits
    print(f"Finished processing file '{file_name}'")

# List of CSV files
csv_files = ['Malameh.csv', 'HM.csv', 'Max Fashion.csv', 'Shein.csv']

# Process each CSV file
for file_name in csv_files:
    process_csv_file(file_name)
    print(f"Completed processing for file '{file_name}'\n")

print("All files processed successfully.")
```

```python
def clean_text(text):
    # Remove mentions (e.g., @username)
    text = re.sub(r'@[A-Za-z0-9_]+', '', text)
    # Remove URLs
    text = re.sub(r'http\S+', '', text)
    # Convert text to lowercase
    text = text.lower()

    return text
```

```python
def getSentiment(comment):
    response = model.generate_content(f"""
        You are a sentiment analysis model. Classify the given Instagram comment into one of the following categories: positive, neutral, or negative. The comment may be in any language. Ensure that every comment receives a classification.

        Respond with only one word: positive, neutral, or negative.

        If the comment is unclear or ambiguous, use your best judgment to classify it. Do not leave any comment unclassified, and do not respond with any words other than positive, neutral, or negative.

        If no comment is provided, respond with neutral.

        Comment: {comment}
    """)
    time.sleep(4)  # Introducing a delay to prevent API rate limits

    # Check if response contains valid parts
    if response and response.parts:
        return response.parts[0].text.strip().lower()
    else:
        # If no valid part is found, return 'neutral'
        print("Invalid response received:")
        print("Comment"+comment)
        print(response)
        return 'neutralInv' #for the comments that the model is unable to deal with
```

```python
csv_files = [f for f in os.listdir() if f.endswith('.csv')]

# Cleaning process and labeling
cleaned_dfs = []
for df_name in csv_files:
    file_name = os.path.splitext(df_name)[0]
    df = pd.read_csv(df_name)

    # Drop unnecessary columns
    df.drop(['post_id', 'post_shortcode', 'commenter_username', 'comment_no_likes', 'post_tagged_users', 'comment_date', 'post_hashtags', 'caption_hashtags'], axis=1, inplace=True)

    # Add Brand column
    df["Brand"] = file_name

    # Normalize Likes
    avg_metrics = get_avg_metrics(file_name)
    if avg_metrics:
        avg_likes, avg_comments, avg_activity = avg_metrics
        df["Normalized_Likes"] = df["post_likes"] / avg_likes
    else:
        df["Normalized_Likes"] = df["post_likes"]

    # Drop post_likes
    df.drop(['post_likes'], axis=1, inplace=True)

    # Clean comment_text using clean_text function
    df["comment_text"] = df["comment_text"].astype(str)
    df["Cleaned_comment"] = df["comment_text"].apply(clean_text)

    # Label sentiment
    df['sentiment'] = df['Cleaned_comment'].apply(getSentiment)

    # Filter to keep only posts and calculate scores
    df_posts = df.drop_duplicates(subset='post_shortcode')
    df_comments = df[~df.index.isin(df_posts.index)]

    # Calculate scores for each post
    df_posts["Negative_Score"] = df_comments[df_comments["sentiment"] == "negative"].groupby("post_shortcode").size() / avg_comments
    df_posts["Positive_Score"] = df_comments[df_comments["sentiment"] == "positive"].groupby("post_shortcode").size() / avg_comments

    df_posts.fillna({"Negative_Score": 0, "Positive_Score": 0}, inplace=True)

    # Append cleaned and labeled DataFrame to cleaned_dfs list
    cleaned_dfs.append(df_posts)

    # Save the cleaned and labeled DataFrame to a CSV file
    df_posts.to_csv(f'Cleaned_and_Labelled_{file_name}.csv', index=False)

    # Wait for 15 minutes before processing the next DataFrame
    print(f"Waiting for 15 minutes before processing the next DataFrame.")
    time.sleep(15 * 60)  # 15 minutes in seconds
```
