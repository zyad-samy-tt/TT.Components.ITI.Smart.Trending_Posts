```python
def convert_to_float(value_str):
    if value_str[-1] == 'k':
        return float(value_str[:-1]) * 1000
    elif value_str[-1] == 'm':
        return float(value_str[:-1]) * 1000000
    else:
        return float(value_str)
```

```python
def get_avg_metrics(brand_name):
    """
    Retrieves the average likes, comments, and activity for a given brand name.

    Args:
    - brand_name (str): The name of the brand to retrieve average metrics for.

    Returns:
    - tuple: A tuple containing average likes, average comments, and average activity for the brand, converted to floats.
    - None: If the brand name is not found in the fashion_brands dictionary.
    """
    # Convert brand_name to lowercase
    brand_name = brand_name.lower()

    # Initialize a variable to store the found brand
    found_brand = None

    # Check if the brand_name matches any key in fashion_brands dictionary in a case-insensitive manner
    for brand in fashion_brands:
        if brand.lower().startswith(brand_name):
            found_brand = brand
            break

    if found_brand:
        # Get the average likes, comments, and activity for the brand
        avg_likes_str = fashion_brands[found_brand]["Stats"].get("Avg. likes", "0")
        avg_comments_str = fashion_brands[found_brand]["Stats"].get("Avg. comments", "0")

        # Convert average likes, comments, and activity strings to float
        def convert_to_float(value_str):
            if value_str[-1] == 'k':
                return float(value_str[:-1]) * 1000
            elif value_str[-1] == 'm':
                return float(value_str[:-1]) * 1000000
            else:
                return float(value_str)

        avg_likes = convert_to_float(avg_likes_str)
        avg_comments = convert_to_float(avg_comments_str)

        return avg_likes, avg_comments
    else:
        return None
```

```python
# Set the working directory
os.chdir('/content/drive/My Drive/Yara/DatasetsWithSA')

# Get a list of all files in the current directory
files = os.listdir()

# Iterate through each file and print average metrics
for file in files:
    if file.endswith('.csv'):
        file_name = os.path.splitext(file)[0]
        print(f"Processing file: {file_name}")
        if file_name.lower() == 'malameh':
          df= pd.read_csv(file, encoding='utf-8', lineterminator='\n')
        else:
          df = pd.read_csv(file)
        print(f"{file}: {df.shape}")
        metrics = get_avg_metrics(file_name)
        if metrics:
            avg_likes, avg_comments = metrics
            print(f"Average metrics for {file_name}:")
            print(f"  Likes: {avg_likes}")
            print(f"  Comments: {avg_comments}")
        else:
            print(f"{file_name} not found in fashion_brands dictionary.")
```

```python
def calculate_sentiment_scores(df, avg_comments):
    print("Calculating sentiment scores...")
    df["comment_count"] = df["comments"].apply(lambda x: len(eval(x)) if isinstance(x, str) else 0)

    print(f"Number of posts: {df.shape[0]}")
    print(f"Total number of comments: {df['comment_count'].sum()}")

    def count_sentiments(sentiments, sentiment_type):
        return sum(1 for sentiment in sentiments if sentiment.lower() == sentiment_type.lower())

    df["positive_count"] = df["sentiment_analysis"].apply(lambda x: count_sentiments(eval(x), "positive") if isinstance(x, str) else 0)
    df["negative_count"] = df["sentiment_analysis"].apply(lambda x: count_sentiments(eval(x), "negative") if isinstance(x, str) else 0)
    df["total_sentiments"] = df["sentiment_analysis"].apply(lambda x: len(eval(x)) if isinstance(x, str) else 0)

    df["positive_score"] = df.apply(lambda row: row["positive_count"] / row["total_sentiments"] if row["total_sentiments"] > 0 else 0, axis=1)
    df["negative_score"] = df.apply(lambda row: row["negative_count"] / row["total_sentiments"] if row["total_sentiments"] > 0 else 0, axis=1)

    return df

def process_csv_files(directory):
    print(f"Setting working directory to: {directory}")
    os.chdir(directory)

    files = os.listdir()
    print(f"Files in directory: {files}")

    cleaned_dfs = []

    for file in files:
        if file.endswith('.csv'):
            file_name = os.path.splitext(file)[0]
            print(f"\nProcessing file: {file_name}")
            if file_name.lower() == 'malameh':
              df= pd.read_csv(file, encoding='utf-8', lineterminator='\n')
            else:
              df = pd.read_csv(file)
            print(f"Loaded {df.shape[0]} rows and {df.shape[1]} columns from {file_name}.csv")

            df["Brand"] = file_name
            print(f"Added 'Brand' column with value '{file_name}'")

            avg_metrics = get_avg_metrics(file_name)
            if avg_metrics:
                avg_likes, avg_comments = avg_metrics
                df["Normalized_Likes"] = df["post_likes"] / avg_likes
                print(f"Normalized likes using average likes: {avg_likes}")

                df = calculate_sentiment_scores(df, avg_comments)
            else:
                df["Normalized_Likes"] = df["post_likes"]
                print("Used original likes (no average metrics found)")

            print("Filtered DataFrame to include required columns")
            df['sentiment_score']=df['positive_score']-df['negative_score']

            cleaned_dfs.append(df)
            print(f"Appended cleaned DataFrame for {file_name}")

    full_df = pd.concat(cleaned_dfs, ignore_index=True)
    print("Concatenated all cleaned DataFrames into one full DataFrame")
    return full_df

# Set the working directory
directory = '/content/drive/My Drive/Yara/DatasetsWithSA'
full_df = process_csv_files(directory)

# Save the merged DataFrame to a CSV file
output_file = 'FullDataset_Cleaned_Labelled_Normalized.csv'
full_df.to_csv(output_file, index=False)
print(f"Saved the final DataFrame to {output_file}")

# Copy the CSV file to Google Drive
os.system(f"cp {output_file} '/content/drive/My Drive/Yara'")
print(f"Copied {output_file} to Google Drive")
```
